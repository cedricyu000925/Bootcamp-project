```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataframe in Tableview CSV.csv")
gender_counts = df.groupby(['Cohort', 'Gender']).size().unstack(fill_value=0)
```

```python
text_representation = df.to_string(index=False)
print(text_representation)
```

```python
gender_counts
```

```python
from IPython.display import Image, display
```

```python
ax = gender_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Gender Distribution by Cohort')
plt.xlabel('Cohort')
plt.ylabel('Number of Students')
plt.legend(title='Gender')
plt.tight_layout()
plt.savefig('gender_distribution.png')
plt.close()
```
